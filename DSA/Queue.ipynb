```python
#class implementation 
class Queue:
    def __init__(self):
        self.buffer=deque() #buffer is an object of type deque
    
    def enqueue(self, val):  #inserting elements into queue 
        self.buffer.appendleft(val)
    
    def dequeue(self): #deleting elements from queue
        if len(self.buffer)==0:
            print("Queue is empty")
            return 
        return self.buffer.pop()

    def is_empty(self):
        return len(self.buffer)==0
    
    def size(self):
        return len(self.buffer)
    
    def front(self):   #returns the element from the front of queue
        return self.buffer[-1]
```

```python
pq = Queue()

pq.enqueue({
    'company': 'Walmart',
    'timestamp': '15 apr, 11.01 AM',
    'price': 131.10
})
pq.enqueue({
    'company': 'Walmart',
    'timestamp': '15 apr, 11.02 AM',
    'price': 132
})
pq.enqueue({
    'company': 'Walmart',
    'timestamp': '15 apr, 11.03 AM',
    'price': 135
})
```

```python
#exercises - code basics dsa playlist - queue 
#question 1 
import threading
import time
from collections import deque
food_order_queue=Queue()

def place_orders(orders):
    for order in orders:
        print("Placing order for:", order)
        food_order_queue.enqueue(order)
        time.sleep(0.5)

def serve_orders():
    time.sleep(1)
    while True:
        order=food_order_queue.dequeue()
        print("Now serving order:", order)
        time.sleep(2)
    

if __name__=='__main__':
    orders = ['pizza','samosa','pasta','biryani','burger']
    t1=threading.Thread(target=place_orders, args=(orders,))
    t2=threading.Thread(target=serve_orders)

    t1.start()
    t2.start()
```
